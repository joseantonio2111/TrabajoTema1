## 1. Un juego de lanzamiento de una moneda

> **1.** Cada jugada del juego consiste en lanzar repetidamente una moneda imparcial hasta que la diferencia entre el número de caras lanzadas y el número de sellos es 3.

> **2.** Si decides jugar el juego, debes pagar 1 por cada lanzamiento de la moneda. No puedes salir durante una jugada del juego.

> **3.** Usted recibe 8 al final de cada juego del juego. Por lo tanto, ganas dinero si el número de lanzamientos requeridos es menor a 8, pero pierdes dinero si se requieren más de 8 lanzamientos.

#### Código

```python
import numpy as np
# Sea 0 cara y 1 sello
tg = []
for j in range(1000):
    c,s,t = 0,0,0
    while abs(s-c) != 3:
        x = np.random.randint(0,2)
        if x == 0:
            c += 1
        else:
            s += 1
        t += 1
    tg.append(t)
    j+=1
```

```python
# Pregunta a
print('La media de los tiros necesarios para ganar es:',np.mean(tg))
print('La desv std de los tiros ganadores es:',np.std(tg))
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(tg,25,alpha=0.7,normed = True)
plt.grid()
plt.xlabel('Tiros ganadores')
plt.show()
```

```python
# Graficando Normal
from scipy import stats as st
mu, sigma = np.mean(tg), np.std(tg)
normal = st.norm(mu, sigma)
x = np.linspace(normal.ppf(0.01),normal.ppf(0.99), 1000)
fp = normal.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
datos = np.random.normal(mu, sigma, 1000) #creando muestra de datos
plt.hist(datos, 20,normed=True)
plt.title('Distribución Normal')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.grid()
plt.show()
```

## 3. Rustbelt Manufacturing Company

>El tiempo requerido por la cuadrilla para reparar una máquina tiene una distribución uniforme durante el intervalo de 0 a dos veces
la media, donde la media depende del tamaño de la cuadrilla. La media es de 4 horas con dos miembros de la cuadrilla, 3 horas con tres miembros de la cuadrilla y 2 horas con cuatro miembros de la cuadrilla.
El tiempo entre las interrupciones de alguna máquina tiene una distribución uniforme de entre 3.5 y 9 horas.

#### Código

```python
# Pregunta a
t = []
p = []
m = 4
for i in range (1000):
    cuad = np.random.uniform(0,2*m)
    esp = np.random.uniform(3.5,9)
    tt = cuad+esp
    t.append(tt)
    i += 1
    if tt > 8:
        p.append(tt)
print('El promedio de tiempo en que una máquina tarda en ser reparada por una cuadrilla de 2 es de',np.mean(t),'horas')
plt.hist(t,normed=True)
plt.title('Histograma')
plt.xlabel('Tiempo')
plt.ylabel('Probabilidad')
plt.grid()
plt.show()
print('La probabilidad que con una cuadrilla de 2 la máquina tarde más de 8 horas en ser reparada es',len(p)/len(t))
```

## 4. Juego de dados

>Él gana si la primera tirada resulta en una suma de 7 u 11 o, alternativamente, si la primera suma es 4, 5, 6, 8, 9 o 10 y la misma suma reaparece antes de que aparezca una suma de 7. Por el contrario, pierde si la primera tirada resulta en una suma de 2, 3 ó 12 o, alternativamente, si la primera suma es 4, 5, 6, 8, 9 o 10 y aparece una suma de 7 antes de la primera suma reaparece.

#### Código

```python
tj = []
d = 500
p = [] # Ganar = 1, perder = 0
for i in range(10000):
    t=0
    d1 = np.random.randint(1,7)
    d2 = np.random.randint(1,7)
    t += 1
    if d1+d2 == 7 or d1+d2 == 11:
        p.append(1)
    elif d1+d2 == 2 or d1+d2 == 3 or d1+d2 == 12:
        p.append(0)
    else:
        boton = d1+d2
        d1 = np.random.randint(1,7)
        d2 = np.random.randint(1,7)
        sd = d1+d2
        t += 1
        while sd != boton and sd != 7:
            d1 = np.random.randint(1,7)
            d2 = np.random.randint(1,7)
            sd = d1+d2
            t += 1
        if sd == boton:
            p.append(1)
        elif sd == 7:
            p.append(0)
    tj.append(t)
    i += 1
```

```python
#Pregunta a
print('La probabilidad de ganar corriendo 10,000 simulaciones es',np.count_nonzero(p)/len(p))
```

```python
#Pregunta b
dt = []
for i in range(10000):
    j=0
    d=500
    while j < 10:
        d1 = np.random.randint(1,7)
        d2 = np.random.randint(1,7)
        t += 1
        if d1+d2 == 7 or d1+d2 == 11:
            d += 50
            j += 1
            dt.append(d)
        elif d1+d2 == 2 or d1+d2 == 3 or d1+d2 == 12:
            d -= 50
            j += 1
            dt.append(d)
        else:
            boton = d1+d2
            d1 = np.random.randint(1,7)
            d2 = np.random.randint(1,7)
            sd = d1+d2
            while sd != boton and sd != 7:
                d1 = np.random.randint(1,7)
                d2 = np.random.randint(1,7)
                sd = d1+d2
            if sd == boton:
                d += 50
            elif sd == 7:
                d -= 50
            dt.append(d)
            j += 1
    i += 1
```

## 5. Inversión CETES

>Una inversión en cetes actualmente promete un rendimiento que va entre el 8 % y el 12 % anual.

```python
d=[]
capital = 300000
for n in range(10000):
    dt = capital*np.random.uniform(1.08,1.1200001)
    d.append(dt)
    n += 1
```

```python
# Pregunta a
print('La media del capital al final del año es',np.mean(d))
print('La desviación estándar del capital al final del año es',np.std(d))
plt.hist(d,normed=True)
plt.title('Histograma')
plt.xlabel('Capital')
plt.grid()
plt.ylabel('Probabilidad')
plt.show()
```

```python
# Pregunta b
total=[]
p58=[]
p55=[]
for x in range(10000):
    capital=300000
    dt=capital*np.random.uniform(1.08,1.1200001)
    for i in range(5):
        dt=dt*np.random.uniform(1.08,1.1200001)
        i+=1
    x+=1
    total.append(dt)
for t in range(len(total)):
    if total[t] - 300000 > 58000:
        p58.append(total[t])
    elif total[t] - 300000 < 55000:
        p55.append(total[t])
    t+=1
```

```python
print('La media del ganancias al final de 6 años es',np.mean(total))
print('La desviación estándar del capital al final de 6 años es',np.std(total))
plt.hist(d,normed=True)
plt.title('Histograma')
plt.xlabel('Capital')
plt.ylabel('Probabilidad')
plt.grid()
plt.show()
print('La probabilidad de tener ganancias > 58000 en 6 años es',len(p58)/len(total))
print('La probabilidad de tener ganancias < 55000 en 6 años es',len(p55)/len(total))
```
